```python
"""
Purpose
-------
Download and plot the last 7 trading days (by minute) values for certain stock 
choices and then plot their values
"""
```

```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# defining the companies
tickers = [
    'AAPL', # Apple Ince
    'MSFT', # Microsoft
    'GOOGL', # Alphabet
    'AMZN', #amazon
    'TSLA', # Tesla
    #'BRK.B', #Berkshire Hathaway
    'NVDA', #nvidia
    'META', #meta
    'TSM', #TAIWAN semiconductor manufacturing
    'V', # Visa
]
```

```python
# downloading the data
period = '5d'
interval = '1m'

df = yf.download(tickers,period = period, interval = interval)
df
```

```python
relative_changes = {}

for ticker in tickers:
    open_price = df["Open"][ticker][0]
    close_price = df["Close"][ticker]
    relative_changes[ticker] = (close_price - open_price)/open_price*100
```

```python
# Plot the relative changes
plt.figure(figsize=(14, 7))

for ticker, relative_change in relative_changes.items():
    plt.plot(relative_change, label=ticker)

plt.title('Relative Change in Stock Prices Over the Last 7 Days')
plt.xlabel('Time')
plt.ylabel('Percent Change')
plt.legend()
plt.grid(True)
plt.show()
```
