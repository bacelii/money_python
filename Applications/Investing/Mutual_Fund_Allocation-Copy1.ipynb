```python
class IndexFund:
    def __init__(
        self,
        index_funds,
        money = 0,
        *args,
        **kwargs):
        super().__init__()

        for k,v in kwargs:
            setattr(self,k,v)
        self.index_funds = index_funds

        self.money = money

    def __str__(self):
        return "\n".join([f"{k}:{v}" for k,v in self.index_funds.items()])

    def money_per_index_fund(self):
        return {k:self.money*v for k,v in self.index_funds.items()}


        
index_fund_allocation = dict(
        economy_index = 0.4,
        large_cap_growth = 0.2,
        total_market = 0.1,
        international = 0.1,
        emerging_markets = 0.1,
        total_bond = 0.1
)

fund = IndexFund(
    index_funds = index_fund_allocation,
    money = 39_557.21
)

fund.money_per_index_fund()
```
